```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv("store_data.csv")
store_data.head(7)
```

```python
records = []
for i in range (0, store_data.shape[0]):
    records.append([str(store_data.values[i, j]) for j in range (0, store_data.shape[1])])
print(records)
```

```python
association_rules = apriori(records, min_support = 0.0045, min_confidence = 0.2, min_lift=3, min_length=2)
#print(len(association_rules))

association_results = list(association_rules)
```

```python
print(association_results[0])
print("\n")
print(association_results[1])
print("\n")
print(association_results[2])
print("\n")
print(association_results[3])
print("\n")
print(association_results[4])
print("\n")
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=========================================")
```
