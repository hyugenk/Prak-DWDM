```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
GroceryStore = pd.read_csv("GroceryStoreDataSet.csv", names = ["product"], header=None)
GroceryStore
```

```python
records = list(GroceryStore["product"].apply(lambda x: x.split(",")))
print(records)
```

```python
association_rules = apriori(records, min_support = 0.003, min_confidence = 0.2, min_lift=3, min_length=2)
#print(len(association_rules))

association_results = list(association_rules)
```

```python
print(association_results[0])
print("\n")
print(association_results[1])
print("\n")
print(association_results[2])
print("\n")
print(association_results[3])
print("\n")
print(association_results[4])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=========================================")
```

#### **Penjelasan**
Perbedaan hasil dari pengolahan dataset aturan asosiasi dengan nilai minimum support dan minimum confidence yang berbeda dapat memiliki dampak
signifikan terhadap aturan yang dihasilkan. Mari kita jelaskan perbedaan tersebut:
#### Minimum Support 0,3% dan Minimum Confidence 20%:
1.	Banyaknya Aturan:
    Karena minimum support lebih rendah (0,3%), lebih banyak transaksi atau itemset akan dianggap signifikan, dan ini dapat menghasilkan lebih
    banyak aturan asosiasi.
2.	Inklusivitas:
    Aturan yang dihasilkan mungkin lebih inklusif, mencakup hubungan yang lebih umum atau kurang kuat antara item.
3.	Sensitivitas terhadap Data Noise:
    Dengan support yang rendah, model cenderung lebih sensitif terhadap noise dalam data, mungkin menghasilkan aturan yang tidak begitu 
    dapat diandalkan.
4.	Jumlah Itemset yang Terlibat:
    Kemungkinan melibatkan lebih banyak itemset, yang bisa meningkatkan kompleksitas aturan.
#### Minimum Support 0,7% dan Minimum Confidence 60%:
1.	Keakuratan dan Keterandalan:
    Karena minimum support dan minimum confidence lebih tinggi, aturan yang dihasilkan mungkin lebih akurat dan dapat diandalkan. 
    Hanya aturan dengan dukungan dan kepercayaan yang tinggi yang akan dipertimbangkan.
2.	Spesifikitas Aturan:
    Aturan yang dihasilkan mungkin lebih spesifik dan hanya melibatkan item atau transaksi yang benar-benar signifikan dalam dataset.
3.	Reduksi Aturan yang Tidak Signifikan:
    Karena ambang batas yang lebih tinggi, aturan yang dihasilkan cenderung lebih terbatas, dan aturan yang tidak memenuhi standar tinggi 
    akan dieliminasi.
4.	Efisiensi Perhitungan:
    Kinerja komputasional mungkin lebih efisien karena jumlah aturan yang dihasilkan lebih sedikit.
